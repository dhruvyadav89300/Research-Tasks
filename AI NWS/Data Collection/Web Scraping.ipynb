```python
url = f"https://forecast.weather.gov/product.php?site=EWX&issuedby=EWX&product=AFD&format=CI&version={version}&glossary=1"
text_data = []
for i in range(1, 45):
    version = i
    page = requests.get(url)
    print(page)
    print("\n")
    print(f"Scraping Version: {version}")
    text = soup.find('pre', class_="glossaryProduct")
    if text != None:
        print("Text fetched succesfully.\nNow attempting storing....")
    else:
        print("Frak")
    text_data.append(text.text)
    print(f"Stored Version {version} successfully")
```

```python
df = pd.DataFrame(text_data)
df.columns = ["Text"]
```
